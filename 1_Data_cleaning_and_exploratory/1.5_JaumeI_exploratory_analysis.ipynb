## Autocorrelation and decomposition of the time series

```python
# People traffic autocorrelation plot

from pandas.plotting import autocorrelation_plot
plt.figure(figsize=(20, 10))
autocorrelation_plot(jaume_df['TotalEntries'])
plt.show()
```

The same weekly cycle previously seen in the other time series can be observed as well in Jaume I.

```python
def append_lags(x, N):
    temp_df = x.copy()
    for i in range (1, N+1):
        temp_df['t-' + str(i)] = x['TotalEntries'].shift(i)
    return temp_df.iloc[N:]

# Plotting correlations creating a month of lags:

jaume_lags = append_lags(jaume_df, (48*7*4))
plt.figure(figsize=(20, 10))
autocor1m = jaume_lags.drop(['Temperature', 'Precipitation', 'Open/Closed'], axis=1).corr()['TotalEntries'].plot(kind='bar')
autocor1m.axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# Finally, let's decompose the time series assuming the seasonal component is additive and a frequency on 1 week:

jaume_ddec = seasonal_decompose(jaume_df['TotalEntries'].astype(np.float64), freq=(48*7))
jaume_ddec.plot()
plt.show()
```

### Outliers

```python
jaume_df['TotalEntries'].sort_values(ascending=False).head(10)
```

```python
jaume_df['TotalEntries'].plot(kind='box')
plt.title('Box plot of TotalEntries of the whole time series', fontsize=15)
plt.show()
```

The boxplot shows quite a few outliers. By taking a look at the line plot, the distribution of TotalEntries along the series doesn't show any noticeable peak. One reason that could explain such a big amount of outliers could be that the series is relatively small (specially if compared with the Maragall time series). If it were longer, some of the observations currently considered as outliers by the boxplot would probably become more frequent. Following this hypothesis, we will not manipulate the outliers of the time series.
