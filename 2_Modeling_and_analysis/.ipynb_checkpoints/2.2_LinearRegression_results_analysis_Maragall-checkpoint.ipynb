```python
# Let's extract the results at the original frequency, daily frequency and their runtime data frames:

OF_results_df = pd.read_csv('./../../../Databases/results_data/lr_results.csv', index_col = 'Timestamp')
DF_results_df = pd.read_csv('./../../../Databases/results_data/lr_dailyresults.csv', index_col = 'Timestamp')
OF_runtime_df = pd.read_csv('./../../../Databases/results_data/lr_runtime_scenarios.csv')
DF_runtime_df = pd.read_csv('./../../../Databases/results_data/lr_runtime_dailyscenarios.csv')
```

## Original frequency anaylsis

The analysis will be performed on the results extracted from the Maragall data, because is the one with most observations and the hypothesis is that this will serve to stablish the basis from which to extract conclusions to be applied on the other time series.

```python
OF_results_df.head()
```

```python
get_scoring_measures(OF_results_df, 336, 340)
```

As expected, using future known observations as lags yields a much better result than using the predictions made by the model as lags. There's a noticeable difference on the RMSE using just lags or using the extra features and external information as well, which leads to better results.

The R2-adjusted ($\bar{R}^2$) score is estimated as a way to infere the amount of variance explained by our model. We need the number of independent regressors for each scenario: in all_cols 340 regressors have been used, while on just_lags, 336.

```python
# Let's plot just a few observations to see the differences between scenarios:

ax = OF_results_df[['Observed_values', 'TotEntr_PKF_allcols', 'TotEntr_PKF_lags', 'TotEntr_PUF_allcols', 'TotEntr_PUF_lags']][:100].plot(figsize=(20,10), alpha=0.8, legend=True)
plt.show()
```

On the control scenarios (PKF, Predicting Known Future, when lags of observed values are used as regressors), using all the columns (including external data) doesn't seem to make a big difference (as observed in the plots and the results stats). On the other hand, when predictions are used each iteration to create new lags for the next prediction, using external information leads to better results.

The results lead to the conclusion that the PUF_allcols scenario (Predicting Unknow Future, including all cols) is the best to perform predictions without having future data. We will continue our analysis using the PUF_allcols predictions.

### Residuals analysis

```python
# Let's plot the residuals to see if there's a visible pattern:

lr_residuals = OF_results_df['Observed_values'] - OF_results_df['TotEntr_PUF_allcols']

lr_residuals.plot(figsize=(20,10))
plt.show()
```

```python
lr_residuals.describe()
```

The description shows an average bias of -13.34 and the bulk of the residuals is on the negative values. A density plot will confirm this.

```python
# Histogram and density of residuals

ax = lr_residuals.hist()
lr_residuals.plot(kind='kde', ax=ax, secondary_y=True, xlim=(-400,400))
plt.title('Distribution of Linear Regression Residuals', fontsize=15)
plt.show()
```

```python
# Autocorrelation plot

from statsmodels.graphics.tsaplots import plot_acf
with plt.rc_context():
    plt.rc("figure", figsize=(20,10))
    plot_acf(lr_residuals)
```

A clear, cyclical pattern can be observed on the autocorrelation plot of the residuals. The forecasts from a model with autocorrelated errors are still unbiased, and so are not "wrong", but they will usually have larger prediction intervals than they need to. [1]

```python
# Plotting residuals against fitted values:

plt.figure(figsize=(20,10))
plt.scatter(OF_results_df['Observed_values'], lr_residuals, alpha=0.05)
plt.show()
```

There seems to be a positive trend, where bigger observed values are connected to bigger positive residual values. There may be "heterocedasticity" in the errors which means the variance of the residuals may not be constant [2] This fact suggests that there might a benefit in applying a transformation of the forecast variable (using for example a Box-Cox transform).

```python
# Let's estimate the RMSE scoring when predicting time windows of increasing span:


lr_future_scoring = pd.DataFrame

for i in [1, 3, 7, 10, 12, 15]:
    if lr_future_scoring.empty:
        lr_future_scoring = pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(OF_results_df['Observed_values'][:48], OF_results_df['TotEntr_PUF_allcols'][:48])), name = (str(i) + '-day')))
        
    else:
        lr_future_scoring = pd.concat([lr_future_scoring, pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(OF_results_df['Observed_values'][:48*i], OF_results_df['TotEntr_PUF_allcols'][:48*i])), name = (str(i)) + '-day'))], axis=1)

lr_future_scoring = pd.DataFrame(lr_future_scoring.values.reshape(-1, 1),
                                index=list(lr_future_scoring.columns),
                                columns=['lr-RMSE'])

lr_future_scoring
```

The time window that seems to minimize the RMSE is the first 7 days: the explanation is that in a 7 day window, the errors of the forecast (the difference between predictions and observed values) on average are smaller than when considering shorter or longer periods.

```python
# Let's plot the distribution of absolute errors along the first week

ax = abs(OF_results_df['Observed_values'] - OF_results_df['TotEntr_PUF_allcols'])[:48*7].plot(figsize=(20,10), alpha=0.5, title='Distribution of absolute errors along the first 7 days and MAE')
plt.plot([mean_absolute_error(OF_results_df['Observed_values'][:48*7], OF_results_df['TotEntr_PUF_allcols'][:48*7]) for e in range(48*7)], color='red')
plt.show()
```

The first day of estimations show big errors, a tendency that fades away with passing days.

```python
# Plotting observed vs. predicted along the first week

ax = OF_results_df[['Observed_values', 'TotEntr_PUF_allcols']][:48].plot(figsize=(20,10), alpha=0.5, legend=True)
OF_results_df[['Observed_values', 'TotEntr_PUF_allcols']][:48*3].plot(figsize=(20,10), alpha=0.5, legend=False, ax = ax)
OF_results_df[['Observed_values', 'TotEntr_PUF_allcols']][:48*7].plot(figsize=(20,10), alpha=0.5, legend=False, ax = ax)

plt.show()
```

The first day of observations shows a big difference from the observed values, thus resulting in a big RMSE. When predicting more values, it seems they adapt better to the observations, diluding the error (since it's calculated as a mean).

## Daily frequency anaylsis

```python
DF_results_df.head(10)
```

```python
# Allcols regressors: 10     Justlags regressors: 7

get_scoring_measures(DF_results_df, 7, 10)
```

As can be observed, the RMSE is bigger than on the 48 observations / day scenarios, since each observation per day is bigger (because of the sum aggregation). The R2_adjusted using future information is lower than on the high frequency models. The negative values can be interpreted as zero.

The explanation is that models with negative values don't fit the data. Let's create a few plots to get a better undestanding of this models.

```python
# Let's plot just a few observations to see the differences between scenarios:

ax = DF_results_df[list(DF_results_df.columns)].plot(figsize=(20,10), alpha=0.8, legend=True)
plt.show()
```

The PKF_allcols (Predicting Known Future, using lags extracted from future observations combined with external data) is the model that better adjusts to the actual observed values. As can be seen, with the passing days the signal captured by the unknown future models fades away until is lost. That's most noticeable on the model that just uses lags: on the PUF_allcols there's a repetitive signal still shown (probably from the features that are not lags).

There's one thing worth mentioning: one of our regressors on the high frequency data was 'Open/closed', feature lost when grouping by day. Maybe it has a big impact on predictions that does not apply to the daily scenario.

[1][2] Hyndman, R.J. and Athanasopoulos, G. (2018) Forecasting: Principles and Practice, 2nd Ed. 

## Comparison of results by resolution

```python
# Grouping predictions by day: comparing observed vs. 48 obs/day vs. daily observations.

obs_preds_grouped = OF_results_df['Observed_values']
obs_preds_grouped.index = pd.to_datetime(obs_preds_grouped.index)
obs_preds_grouped = obs_preds_grouped.groupby(pd.Grouper(freq='D')).sum()

est_preds_grouped = OF_results_df['TotEntr_PUF_allcols']
est_preds_grouped.index = pd.to_datetime(est_preds_grouped.index)
est_preds_grouped = est_preds_grouped.groupby(pd.Grouper(freq='D')).sum()

grouped_predictions = pd.concat([pd.DataFrame(pd.Series(obs_preds_grouped, name='obs_vals')), pd.DataFrame(pd.Series(est_preds_grouped, name='est_preds_48/day'))], axis=1)
grouped_predictions = pd.concat([grouped_predictions, DF_results_df['DayEntr_PUF_allcols']], axis=1)
grouped_predictions.head(10)
```

```python
get_scoring_measures(grouped_predictions, 340, 10, comparing_freqs=True)
```

Comparing the scorings of the 2 methods (estimate 48 observations per day and aggregate them vs. making daily predictions), the RMSE is significantly smaller making predictions keeping a high frequency and aggregating them afterwards. Let's compare scorings in different time windows:

```python
# Estimating the total RMSE for windows of diferent span:

for x in [1, 3, 7, 10]:
    print(get_scoring_measures(grouped_predictions[:x], 340, 7, comparing_freqs=True), '     {} days'.format(str(x)))
    print('-'*50)
```

```python
# Estimating the total RMSE for specific days in the future:

for x in [1, 3, 7, 10]:
    print(get_scoring_measures(grouped_predictions[x-1:x], 336, 7, comparing_freqs=True), '     day {}'.format(str(x)))
    print('-'*50)
```

```python
RMSE_comparison_df = pd.DataFrame
RMSE_48_df = pd.DataFrame 
RMSE_daily = pd.DataFrame



for i in range(1,15):
    if RMSE_48_df.empty:
        RMSE_48_df = pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(grouped_predictions['obs_vals'][:1], grouped_predictions['est_preds_48/day'][:1])), name = 'RMSE-48')) 
        RMSE_daily = pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(grouped_predictions['obs_vals'][:1], grouped_predictions['DayEntr_PUF_allcols'][:1])), name = 'RMSE-daily'))
        
    else:
        RMSE_48_df = pd.concat([RMSE_48_df, pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(grouped_predictions['obs_vals'][i-1:i], grouped_predictions['est_preds_48/day'][i-1:i])), name = 'RMSE-48'))])
        RMSE_daily = pd.concat([RMSE_daily, pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(grouped_predictions['obs_vals'][i-1:i], grouped_predictions['DayEntr_PUF_allcols'][i-1:i])), name = 'RMSE-daily'))])
       
        
RMSE_comparison_df = pd.concat([RMSE_48_df, RMSE_daily], axis = 1)
RMSE_comparison_df.index = obs_preds_grouped[:len(RMSE_daily)].index

RMSE_comparison_df
```

```python
# Let's plot the RMSE of the 2 models (daily and 48 predictions) of the first 14 days:

ax = RMSE_comparison_df.plot(figsize=(20,10), legend=True)
ax.tick_params(labelsize=20)
ax.axes.tick_params(which='minor', labelsize=20)
ax.xaxis.label.set_fontsize(20)
ax.legend(loc=2, prop={'size': 20})

plt.show()
```

In general terms, for the first 14 days the RMSE showed by the model using 48 preds./day seems smaller.

## Computing cost

```python
OF_runtime_df.columns, DF_runtime_df.columns
```

```python
# Let's create a unique data frame containing all computing costs

DF_runtime_df.columns = OF_runtime_df.columns

OF_runtime_df = pd.DataFrame(OF_runtime_df.values.reshape(-1, 1),
                              index=list(OF_runtime_df.columns),
                              columns=['Original_freq'])

DF_runtime_df = pd.DataFrame(DF_runtime_df.values.reshape(-1, 1),
                              index=list(DF_runtime_df.columns),
                              columns=['Daily_freq'])

runtime_df = pd.concat([OF_runtime_df, DF_runtime_df], axis = 1)
runtime_df
```

```python
# The units are seconds: let's translate it to hours dividing by 3600 (60^2)

runtime_df / 3600
```

Surprisingly, on the high frequency scenario, the calculations to obtain the results just using lags are more expensive than using all the columns: it's worth to mention because it argues against intuition. Since using all columns implies an extra operation of concatenating 2 data frames each iteration it could be thought to be more expensive than just calculating lags.

For the linear regression algorithm, even though the cost of computing daily data is much cheaper, the model that seems to better capture the signal of the time series is the one that uses all columns and high frequency data.
