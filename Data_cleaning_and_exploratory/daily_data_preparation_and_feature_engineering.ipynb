```python
import numpy as np
import pandas as pd
import datetime
```

```python
traffic_df = pd.read_excel('./../../../Databases/data.xlsx')
bcn_meteo_df = pd.read_csv('./../../../Databases/X4Barcelona_converted.csv')
can_meteo_df = pd.read_csv('./../../../Databases/XUCanyelles_converted.csv')
```

```python
bcn_meteo_df.head()
```

```python
# Downsampling the meteo data

bcn_meteo_df['Timestamp'] = pd.to_datetime(bcn_meteo_df['Timestamp'], dayfirst=True)
can_meteo_df['Timestamp'] = pd.to_datetime(can_meteo_df['Timestamp'], dayfirst=True)

bcn_meteo_df.set_index('Timestamp', inplace=True)
bcn_meteo_df.sort_index(inplace=True)
bcn_daily_meteo = bcn_meteo_df.resample('D', how='mean')

can_meteo_df.set_index('Timestamp', inplace=True)
can_meteo_df.sort_index(inplace=True)
can_daily_meteo = bcn_meteo_df.resample('D', how='mean')
```

```python
bcn_daily_meteo.head()
```

```python
# Downsampling the traffic data by location

ravn_daily = traffic_df[traffic_df['SiteId'] == 1][['Timestamp', 'TotalEntries']]
ravn_daily.sort_values(['Timestamp'], ascending=True, inplace=True)
ravn_daily['Timestamp'] = pd.to_datetime(ravn_daily['Timestamp'], unit='s')
ravn_daily.set_index('Timestamp', inplace=True)
ravn_daily.sort_index(inplace=True)
ravn_daily = ravn_daily.resample('D', how='sum')

marag_daily = traffic_df[traffic_df['SiteId'] == 2][['Timestamp', 'TotalEntries']]
marag_daily.sort_values(['Timestamp'], ascending=True, inplace=True)
marag_daily['Timestamp'] = pd.to_datetime(marag_daily['Timestamp'], unit='s')
marag_daily.set_index('Timestamp', inplace=True)
marag_daily.sort_index(inplace=True)
marag_daily = marag_daily.resample('D', how='sum')

jaume_daily = traffic_df[traffic_df['SiteId'] == 3][['Timestamp', 'TotalEntries']]
jaume_daily.sort_values(['Timestamp'], ascending=True, inplace=True)
jaume_daily['Timestamp'] = pd.to_datetime(jaume_daily['Timestamp'], unit='s')
jaume_daily.set_index('Timestamp', inplace=True)
jaume_daily.sort_index(inplace=True)
jaume_daily = jaume_daily.resample('D', how='sum')

marag_daily.head()
```

```python
# Appending traffic and meteo data by location

ravn_daily = ravn_daily.merge(bcn_daily_meteo.drop(['TX', 'TN'], axis = 1), how='left', on='Timestamp')
marag_daily = marag_daily.merge(bcn_daily_meteo.drop(['TX', 'TN'], axis = 1), how='left', on='Timestamp')
jaume_daily = jaume_daily.merge(can_daily_meteo.drop(['TX', 'TN'], axis = 1), how='left', on='Timestamp')

ravn_daily.sort_index(inplace=True)
marag_daily.sort_index(inplace=True)
jaume_daily.sort_index(inplace=True)

marag_daily.head()
```

#### Working_days/Weekend_days

```python
ravn_daily['weekday'] = ravn_daily.index.weekday
ravn_daily['Weekday/Weekend'] = 1
ravn_daily['Weekday/Weekend'][ravn_daily['weekday'] > 4] = 0
ravn_daily.drop(['weekday'], axis=1, inplace=True)

marag_daily['weekday'] = marag_daily.index.weekday
marag_daily['Weekday/Weekend'] = 1
marag_daily['Weekday/Weekend'][marag_daily['weekday'] > 4] = 0
marag_daily.drop(['weekday'], axis=1, inplace=True)

jaume_daily['weekday'] = jaume_daily.index.weekday
jaume_daily['Weekday/Weekend'] = 1
jaume_daily['Weekday/Weekend'][jaume_daily['weekday'] > 4] = 0
jaume_daily.drop(['weekday'], axis=1, inplace=True)

marag_daily.head(10)
```

####  Eliminating Maragall discontinuous data

```python
marag_daily = marag_daily[5:]
marag_daily.head()
```

#### Creating lags for each dataframe

```python
def append_lags(x, N):
    temp_df = x.sort_index().copy()
    for i in range (1, N+1):
        temp_df['t-' + str(i)] = x['TotalEntries'].shift(i)
    return temp_df.iloc[N:]
```

```python
ravn_lags_daily = append_lags(ravn_daily, 7)
marag_lags_daily = append_lags(marag_daily, 7)
jaume_lags_daily = append_lags(jaume_daily, 7)

marag_lags_daily.head()
```

```python
ravn_daily.to_csv('./../../../Databases/ravaln_daily_data.csv', index=True)
marag_daily.to_csv('./../../../Databases/marag_daily_data.csv', index=True)
jaume_daily.to_csv('./../../../Databases/jaume_daily_data.csv', index=True)

ravn_lags_daily.to_csv('./../../../Databases/ravaln_dailylags_data.csv', index=True)
marag_lags_daily.to_csv('./../../../Databases/marag_dailylags_data.csv', index=True)
jaume_lags_daily.to_csv('./../../../Databases/jaume_dailylags_data.csv', index=True)
```
