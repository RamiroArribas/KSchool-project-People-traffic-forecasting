### Residuals analysis

```python
# Let's plot the residuals to see if there's a visible pattern:

lr_residuals = OF_results_df['Observed_values'] - OF_results_df['TotEntr_PUF_allcols']

lr_residuals.plot(figsize=(20,10))
plt.show()
```

```python
lr_residuals.describe()
```

The description shows an average bias of -13.34 and bulk of the residuals is on the negative values. A density plot will confirm this.

```python
# Histogram and density of residuals

ax = lr_residuals.hist()
lr_residuals.plot(kind='kde', ax=ax, secondary_y=True, xlim=(-200,200))
plt.title('Distribution of Linear Regression Residuals', fontsize=15)
plt.show()
```

```python
# Autocorrelation plot

from statsmodels.graphics.tsaplots import plot_acf
with plt.rc_context():
    plt.rc("figure", figsize=(20,10))
    plot_acf(lr_residuals)
```

A clear, cyclical pattern can be observed on the autocorrelation plot of the residuals. The forecasts from a model with autocorrelated errors are still unbiased, and so are not "wrong", but they will usually have larger prediction intervals than they need to. [1]

```python
# Plotting residuals against fitted values:

plt.figure(figsize=(20,10))
plt.scatter(OF_results_df['Observed_values'], lr_residuals, alpha=0.05)
plt.show()
```

There seems to be a positive trend, where bigger observed values are connected to bigger positive residual values. There may be "heterocedasticity" in the errors which means the variance of the residuals may not be constant [2] This fact suggests that there might a benefit in applying a transformation of the forecast variable (using for example a Box-Cox transform).

```python
# Let's estimate the scoring when predicting different time windows of days into the future:


lr_future_scoring = pd.DataFrame

for i in [1, 3, 7, 10, 12, 15]:
    if lr_future_scoring.empty:
        lr_future_scoring = pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(OF_results_df['Observed_values'][:48], OF_results_df['TotEntr_PUF_allcols'][:48])), name = (str(i) + '-day')))
        
    else:
        lr_future_scoring = pd.concat([lr_future_scoring, pd.DataFrame(pd.Series(np.sqrt(mean_squared_error(OF_results_df['Observed_values'][:48*i], OF_results_df['TotEntr_PUF_allcols'][:48*i])), name = (str(i)) + '-day'))], axis=1)

lr_future_scoring = pd.DataFrame(lr_future_scoring.values.reshape(-1, 1),
                                index=list(lr_future_scoring.columns),
                                columns=['lr-RMSE'])


lr_future_R2_df = pd.DataFrame

for i in [1, 3, 7, 10, 12, 15]:
    if lr_future_R2_df.empty:
        lr_future_R2_df = pd.DataFrame(pd.Series(r2_score(OF_results_df['Observed_values'][:48], OF_results_df['TotEntr_PUF_allcols'][:48]), name = (str(i) + '-day')))
        
    else:
        lr_future_R2_df = pd.concat([lr_future_R2_df, pd.DataFrame(pd.Series(r2_score(OF_results_df['Observed_values'][:48*i], OF_results_df['TotEntr_PUF_allcols'][:48*i]), name = (str(i) + '-day')))], axis=1)
        
lr_future_R2_df = pd.DataFrame(lr_future_R2_df.values.reshape(-1, 1),
                              index=list(lr_future_R2_df.columns),
                              columns=['R2'])

lr_future_R2_df['lr_R2_adj'] = (1 - (1 - lr_future_R2_df['R2'])) * ((OF_results_df.shape[0] - 1)/(OF_results_df.shape[0] - 340 - 1))
lr_future_scoring = pd.concat([lr_future_scoring, lr_future_R2_df['lr_R2_adj']], axis=1)
lr_future_scoring
```

The time window that seems to minimize the RMSE is the first 7 days: the explanation is that in a 7 day window, the errors of the forecast (the difference between predictions and observed values) on average is less than when considering shorter or longer periods. It doesn't mean that the predictions are better per se, it just reflexes that on average the predictions are more accurated.

```python
# Let's plot the distribution of absolute errors along the first week

ax = abs(OF_results_df['Observed_values'] - OF_results_df['TotEntr_PUF_allcols'])[:48*7].plot(figsize=(20,10), alpha=0.5, title='Distribution of absolute errors along the first 7 days and MAE')
plt.plot([mean_absolute_error(OF_results_df['Observed_values'][:48*7], OF_results_df['TotEntr_PUF_allcols'][:48*7]) for e in range(48*7)], color='red')
plt.show()
```

The first day of estimations show big errors, a tendency that fades away with passing days.

```python
# Plotting observed vs. predicted along the first week

ax = OF_results_df[['Observed_values', 'TotEntr_PUF_allcols']][:48].plot(figsize=(20,10), alpha=0.5, legend=True)
OF_results_df[['Observed_values', 'TotEntr_PUF_allcols']][:48*3].plot(figsize=(20,10), alpha=0.5, legend=False, ax = ax)
OF_results_df[['Observed_values', 'TotEntr_PUF_allcols']][:48*7].plot(figsize=(20,10), alpha=0.5, legend=False, ax = ax)

plt.show()
```

The first day of observations shows a big difference from the observed values, thus resulting in a big RMSE. When predicting more values seems to adapt better to the observations, diluding the error (since it's calculated as a mean).

As can be observed, the RMSE is bigger than on the 48 observations / day scenarios, since each observation per day is bigger (because of the sum aggregation). The negative values obtained from the models using each prediction to extract new lags per iteration can be interpreted as zero.

The explanation is that said models don't fit the data. Let's create a few plots to get a better undestanding of this models.

```python
# Let's plot just a few observations to see the differences between scenarios:

ax = DF_results_df[list(DF_results_df.columns)].plot(figsize=(20,10), alpha=0.5, legend=True)
plt.show()
```

The PKF_allcols (Predicting Known Future, using lags extracted from future observations combined with external data) is the model that better adjusts to the actual observed values. As can be seen, with the passing days the signal captured by the unknown future models fades away until is lost. That's most noticeable on the model that just use lags: on the PUK_allcols there's a repetitive signal still shown (from the columns that are not lags).

There's one thing worth mentioning: one of our regressors on the high frequency data was 'Open/closed', feature lost when grouping by day. Maybe it has a big impact on predictions that does not apply to the daily scenario.

## Comparison of results by resolution

```python
# Grouping predictions by day: comparing observed vs. 48 obs/day vs. daily observations.

obs_preds_grouped = OF_results_df['Observed_values']
obs_preds_grouped.index = pd.to_datetime(obs_preds_grouped.index)
obs_preds_grouped = obs_preds_grouped.groupby(pd.Grouper(freq='D')).sum()

est_preds_grouped = OF_results_df['TotEntr_PUF_allcols']
est_preds_grouped.index = pd.to_datetime(est_preds_grouped.index)
est_preds_grouped = est_preds_grouped.groupby(pd.Grouper(freq='D')).sum()

grouped_predictions = pd.concat([pd.DataFrame(pd.Series(obs_preds_grouped, name='obs_vals')), pd.DataFrame(pd.Series(est_preds_grouped, name='est_preds_48/day'))], axis=1)
grouped_predictions = pd.concat([grouped_predictions, DF_results_df['DayEntr_PUF_allcols']], axis=1)
grouped_predictions.head(10)
```

```python
# High frequency scenario: 340 features    Daily predictions: 10 features

get_scoring_measures(grouped_predictions, 340, 10, comparing_resolution=True)
```

Comparing the scorings of the 2 methods (estimate 48 observations per day and aggregate them vs. making daily predictions), even though the R2_adjusted values can be considered 0 in both, the RMSE is significantly smaller making predictions keeping a high frequency and aggregating them afterwards. Let's compare scoring in different time windows:

```python
# Estimating the total RMSE and R2_adjusted for windows of diferent span:

for x in [1, 3, 7, 10]:
    print(get_scoring_measures(grouped_predictions[:x], 340, 7, comparing_resolution=True), str(x)+'-days')
    print('-'*50)
```

```python
# Estimating the total RMSE and R2_adj for specific days in the future:

for x in [1, 3, 7, 10]:
    print(get_scoring_measures(grouped_predictions[x-1:x], 336, 7, comparing_resolution=True), str(x)+'-days')
    print('-'*50)
```

## Computing cost

```python
OF_runtime_df.columns, DF_runtime_df.columns
```

```python
# Let's create a unique data frame containing al computing costs

DF_runtime_df.columns = OF_runtime_df.columns

OF_runtime_df = pd.DataFrame(OF_runtime_df.values.reshape(-1, 1),
                              index=list(OF_runtime_df.columns),
                              columns=['Original_freq'])

DF_runtime_df = pd.DataFrame(DF_runtime_df.values.reshape(-1, 1),
                              index=list(DF_runtime_df.columns),
                              columns=['Daily_freq'])

runtime_df = pd.concat([OF_runtime_df, DF_runtime_df], axis = 1)
runtime_df
```

```python
# The units are seconds: let's translate it to hours dividing by 3600 (60^2)

runtime_df / 3600
```

Surprisingly, the calculations to obtain the results just using lags are more expensive than using all the columns: it's worth to mention because it argues against intuition. Since using all columns implies an extra operation of concatenating 2 data frames each iteration it could be thought to be more expensive than just calculating lags.
